```python
# Missing values treatments by imputing/filling the missing values using mean & median(numerical) , mode(categorical columns) , KNNImputer(numerical columns) ; mean & meadian for continuous , mode for discrete
df.isnull().sum()
for i  in df.select_dtypes(include="number").columns
    df[i].fillna(df[i].median(),inplace = True)
for i  in df.select_dtypes(include="number").columns
    df[i].fillna(df[i].mean(),inplace = True)

from sklearn.impute import KNNImputer
impute=KNNImputer()
for i in df.select_dtypes(include="number").columns
  df[i]=impute.fit_transform(df[[i]])
```

```python
# Firstly to decide whether to do outliers treatment or not, if so then how to do it ?
# CAP the outliers to the upper/lower wiskers
# No outliers treatments for target variables , categorical & discrete variables ; Its only for continuous numerical data
def wisker(col):
  q1,q3=np.percentile(col,[25,75])
  iqr=q3-q1
  lw=q1-1.5*iqr
  uw=q3+1.5*1qr
  return lw,uw
wisker(df['Population'])
for i  in df.select_dtypes(include="number").columns:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,uw,df[i])
    df[i]=np.where(df[i]>lw,uw,df[i])
for i  in df.select_dtypes(include="number").columns:
    sns.boxplot(df[i])
    plt.show()
```
