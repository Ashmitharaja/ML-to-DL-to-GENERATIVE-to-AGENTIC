```python
# ✅ Mini Large Reasoning Model with Debug Logs

# 📚 Knowledge Base
knowledge_base = [
    ("Paris",   "isCapitalOf", "France"),
    ("France",  "isPartOf",    "Europe"),
    ("Berlin",  "isCapitalOf", "Germany"),
    ("Germany", "isPartOf",    "Europe"),
    ("France",  "hasLanguage", "French"),
    ("Germany", "hasLanguage", "German"),
]
print("✔️ Knowledge Base loaded with", len(knowledge_base), "facts")

# 🔗 Reasoning Engine
def explain_chain(start, end, kb):
    visited = set()

    def dfs(current, target, trace):
        if current == target:
            return trace
        visited.add(current)
        for subj, rel, obj in kb:
            if subj == current and obj not in visited:
                result = dfs(obj, target, trace + [(subj, rel, obj)])
                if result:
                    return result
        return None

    return dfs(start, end, [])

# 🗣️ Parser
import re

# 🗣️ Improved Question Parser
def parse_question(question: str):
    q = question.lower().strip()
    q = re.sub(r'[^\w\s]', '', q)  # Removes punctuation like ?, !, .

    if "capital of" in q:
        country = q.split("capital of")[-1].strip().capitalize()
        return ("?x", "isCapitalOf", country)

    elif "continent is" in q:
        country = q.split("continent is")[-1].strip().capitalize()
        return (country, "isPartOf", "?x")

    elif "language in" in q:
        country = q.split("language in")[-1].strip().capitalize()
        return (country, "hasLanguage", "?x")

    return None


# 🧮 Inference Engine
def answer_query(parsed, kb):
    subj, rel, obj = parsed
    print(f"🔍 Looking for match: {parsed}")
    for s, r, o in kb:
        print(f"   👉 checking ({s}, {r}, {o})")
        if subj == "?x" and r == rel and o.lower() == obj.lower():
            return s
        elif obj == "?x" and s.lower() == subj.lower() and r == rel:
            return o
    return None

# ▶️ Ask Function
def ask(question: str):
    print(f"\n❓ Question: {question}")
    parsed = parse_question(question)
    if not parsed:
        print("🤖 Sorry, I couldn't parse your question.")
        return

    result = answer_query(parsed, knowledge_base)
    if result is None:
        print("🤖 I don't know.")
        return

    print("💡 Answer:", result)
    # Reasoning
    start = parsed[0] if parsed[0] != "?x" else result
    end = parsed[2] if parsed[2] != "?x" else result
    chain = explain_chain(start, end, knowledge_base)
    if chain:
        print("🧠 Reasoning Path:")
        for step in chain:
            print("   •", step[0], step[1], step[2])

# 🚀 Try It
ask("What is the capital of France?")
ask("What continent is Germany in?")
ask("What language in France?")
```

3.Core Components of LRM

Reasoning Graphs: Structured logic paths used internally

Symbolic Engine Integration: Includes tools like theorem provers, solvers

Neural-Symbolic Fusion: Combines deep learning + symbolic logic

Multi-hop Inference: Handles questions needing multiple steps to reach an answer

Memory-Augmented Attention: Remembers context, facts, steps

Step-by-step Justification: Provides traceable explanation (unlike black-box LLMs)

4.Types of Reasoning in LRM

Deductive Reasoning: From general rule → specific result

Inductive Reasoning: From patterns → general rule

Abductive Reasoning: From effect → best cause

Analogical Reasoning: Based on structural similarity

Causal Reasoning: Understands cause-effect chains

Counterfactual Reasoning: “What if” logic

Commonsense Reasoning: Practical human assumptions

5.Architecture (High-Level Stack)

Input → Embedding Layer → Memory Controller → Reasoning Core
→ Rule Engine → Output Generator + Explanation Module
Embedding Layer: Contextualizes input (text, code, facts)

Memory Controller: Accesses external/internal memory

Reasoning Core: Applies inference chains

Rule Engine: Uses explicit logic rules and facts

Output Generator: Returns answer + reasoning trace
