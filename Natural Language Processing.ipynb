**Language Modelling**

**Machine Learning (ML) – Feature Extraction**

1. Raw data is collected and preprocessed (e.g., cleaning, normalization).
2. Manual feature extraction is required before training.
3. Feature engineering is based on domain knowledge.
4. Common techniques for feature extraction:

   * For text: TF-IDF, bag-of-words
   * For images: HOG, SIFT
   * For audio: MFCC
5. Extracted features are usually flat and shallow.
6. These features are then used to train traditional ML models like SVM, decision trees, etc.
7. Feature extraction and model training happen as separate steps.
8. ML works best with structured or tabular data.
9. Suitable for small to medium-sized datasets.


---



**Deep Learning (DL) – Feature Extraction**

1. Raw data is directly fed into the deep learning model.
2. Minimal preprocessing is required (e.g., resizing, normalization).
3. No manual feature extraction is needed.
4. The model automatically learns relevant features during training.
5. Deep learning uses multiple layers to learn hierarchical features.
   * Lower layers learn basic patterns (e.g., edges, shapes).
   * Higher layers learn complex features (e.g., objects, meaning).
6. Feature extraction and model training happen simultaneously in an end-to-end process.
7. DL works well with unstructured data like images, audio, and text.
8. Requires large amounts of data and computing resources.



**Features of the text data**

**1. Lexical Features (Surface-level text features)**
These are based on the raw form of the text.
* Word count
* Character count
* Average word length
* Sentence count
* Average sentence length
* Frequency of specific words
* Number of unique words (vocabulary richness)
* Presence of uppercase words
* Frequency of punctuation (e.g., ?, !, ,)
* Presence of digits or special characters

**2. Syntactic Features (Grammar and structure)**
* Part-of-Speech (POS) tags (e.g., noun, verb, adjective)
* POS tag frequency distribution
* N-grams (unigrams, bigrams, trigrams)
* Parse trees (constituency or dependency parsing)
* Sentence complexity or depth
* Use of passive vs. active voice

**3. Semantic Features (Meaning-based)**
* Named entities (persons, organizations, locations)
* Semantic role labeling (who did what to whom)
* Word embeddings (e.g., Word2Vec, GloVe, FastText)
* Sentence embeddings (e.g., BERT, Sentence-BERT)
* Word sense disambiguation
* Synonym and antonym presence
* Coreference resolution (e.g., "he" refers to "John")

**4. Sentiment & Emotion Features**
* Sentiment polarity (positive, negative, neutral)
* Sentiment intensity scores
* Emotion tags (anger, joy, sadness, etc.)
* Subjectivity score (subjective vs. objective text)

**5. Stylistic & Writing Style Features**
* Readability scores (Flesch–Kincaid, Gunning Fog)
* Use of personal pronouns (I, you, we)
* Formality or informality level
* Slang or abbreviation usage
* Spelling or grammatical error count

**6. Statistical & Frequency Features**
* Term Frequency (TF)
* Inverse Document Frequency (IDF)
* TF-IDF values
* Bag of Words representation
* Document frequency of terms across a corpus

**7. Structural / Format Features**
* Paragraph count
* Line breaks or formatting markers
* HTML/XML tags (in web data)
* Hashtags, mentions (in social media data)
* Position of words (beginning, middle, end)

**8. Custom Domain-Specific Features**
* Presence of domain-specific keywords (e.g., in medical, legal, finance texts)
* Use of technical jargon or acronyms
* Text label annotations (for supervised tasks)



**Vocabulary**

1. **Corpus**:

   * A corpus is a large collection of text data (e.g., a set of documents, tweets, or articles).

2. **Tokens**:

   * When you split the corpus into individual words or terms, each of these is called a *token*.

3. **Vocabulary**:

   * The vocabulary is the **set of all unique tokens** in the entire corpus.
   * It removes duplicates and may involve preprocessing (e.g., lowercasing, punctuation removal).

4. **Vocabulary Size**:

   * Refers to the **total number of unique words** in the corpus.
   * Example: If the corpus contains the words: `["dog", "Dog", "cat", "cat"]`
     → After lowercasing: `["dog", "cat"]`
     → Vocabulary = `{"dog", "cat"}`
     → Vocabulary size = 2

5. **Preprocessing Effects**:

   * Techniques like stopword removal, stemming, lemmatization, and case-folding affect the final vocabulary.

6. **Use of Vocabulary**:

   * To create feature representations like Bag of Words, TF-IDF, and one-hot encoding.
   * Forms the base for embedding layers in deep learning models.

7. **Vocabulary in NLP Models**:

   * In traditional ML: defines columns/features in vectorization.
   * In DL: used to build embedding matrices or tokenizers.
